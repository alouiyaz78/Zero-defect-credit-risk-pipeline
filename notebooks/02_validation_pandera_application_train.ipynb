# Phase 1 - Validation automatique avec Pandera (`application_train`)

Ce notebook montre comment passer d'un contrôle qualité exploratoire à une validation automatique des données avec **Pandera**.

## Objectifs

Dans cette étape, on cherche à :

- charger le dataset principal `application_train.csv`
- appliquer un schéma de validation Pandera
- vérifier les règles métier principales
- afficher un aperçu des données validées
- analyser les valeurs manquantes après validation
- illustrer la capture d'erreurs sur un exemple volontairement incorrect

Ce notebook complète le notebook de contrôle qualité exploratoire réalisé précédemment.

## 1. Import des bibliothèques

```python
from pathlib import Path
import sys
import pandas as pd
import warnings

warnings.filterwarnings("ignore")

pd.set_option("display.max_columns", 200)
pd.set_option("display.float_format", lambda x: f"{x:,.4f}")
```

## 2. Chargement du fichier de validation Pandera

```python
PROJECT_ROOT = Path.cwd().parent

if str(PROJECT_ROOT) not in sys.path:
    sys.path.append(str(PROJECT_ROOT))

from configs.pandera_application_train import validate_application_train
```

## 3. Chargement du dataset principal

Le dataset `application_train.csv` est la table principale du projet.  
C'est le seul fichier qui contient la variable cible `TARGET`, donc c'est à partir de cette table que l'entraînement du modèle sera réalisé plus tard.

Les autres fichiers serviront à enrichir ce dataset après agrégation et fusion.

```python
DATA_PATH = PROJECT_ROOT / "data" / "raw" / "application_train.csv"
train = pd.read_csv(DATA_PATH)

print("Chemin du fichier :", DATA_PATH)
print("Dimensions du dataset :", train.shape)

train.head()
```

## 4. Validation du dataset

On applique ici la fonction `validate_application_train()` définie dans le fichier Pandera.

Cette validation permet de vérifier automatiquement plusieurs règles, par exemple :

- `SK_ID_CURR` doit être unique
- `TARGET` doit être binaire (`0` ou `1`)
- les montants financiers doivent être positifs ou nuls
- `DAYS_BIRTH` doit être négatif (codage du dataset)
- `DAYS_EMPLOYED` doit être inférieur ou égal à `0` ou égal à `365243`
- les variables catégorielles doivent appartenir à une liste de valeurs autorisées
- l'âge dérivé doit être compris entre `18` et `100` ans

```python
try:
    validated_train = validate_application_train(train)

    print("Validation réussie")
    print("Dimensions après validation :", validated_train.shape)

    display(
        validated_train[
            ["SK_ID_CURR", "TARGET", "AMT_INCOME_TOTAL", "DAYS_BIRTH", "AGE_YEARS"]
        ].head()
    )

    print("\nValeurs manquantes (top 5) :")
    print(validated_train.isnull().sum().sort_values(ascending=False).head())

except Exception as e:
    print("Une erreur a été détectée lors de la validation des données :")
    print(e)
```

## 5. Interprétation des résultats de validation

Lorsque la validation est réussie, cela signifie que les colonnes contrôlées respectent les règles définies dans le schéma Pandera.

Cela ne veut pas dire que toutes les données sont parfaites, mais cela garantit déjà plusieurs points essentiels :

- la structure minimale du dataset est correcte
- les colonnes clés sont présentes
- les types sont cohérents
- certaines incohérences métier évidentes sont bloquées automatiquement

Cette étape permet donc de sécuriser le pipeline avant le feature engineering et la modélisation.

## 6. Analyse complémentaire des valeurs manquantes

La validation Pandera n'empêche pas la présence de valeurs manquantes sur les colonnes où elles sont autorisées.  
Il est donc utile de compléter la validation automatique par un résumé des taux de valeurs manquantes.

```python
missing_df = (
    validated_train.isnull().mean().mul(100).sort_values(ascending=False).reset_index()
)
missing_df.columns = ["colonne", "pourcentage_manquant"]
missing_df["nb_manquant"] = missing_df["colonne"].map(validated_train.isnull().sum())

missing_df.head(15)
```

### Observation

On remarque que certaines variables, notamment liées aux caractéristiques immobilières, présentent un taux très élevé de valeurs manquantes.

Cela n'est pas forcément une erreur de donnée.  
Il peut s'agir :

- d'informations non disponibles pour tous les clients
- de variables non renseignées dans certains contextes
- de colonnes qui devront être supprimées, imputées ou transformées plus tard

Cette observation sera utile dans la phase de prétraitement.

## 7. Vérification ciblée de l'âge dérivé

```python
validated_train["AGE_YEARS"].describe()
```

### Interprétation

Dans ce dataset, `DAYS_BIRTH` est négatif car il représente un nombre de jours relatif à la date de la demande.  
L'âge réel est donc calculé ainsi :

`AGE_YEARS = - DAYS_BIRTH / 365.25`

La validation vérifie ensuite que cet âge est compris entre `18` et `100` ans.

## 8. Exemple d'erreur capturée par Pandera

```python
try:
    bad_sample = train.copy()
    bad_sample.loc[0, "TARGET"] = 3
    validate_application_train(bad_sample)
except Exception as e:
    print("Erreur capturée avec succès :")
    print(str(e)[:1500])
```

## 9. Exemple d'erreur métier complémentaire

Dans cette partie, on provoque volontairement une incohérence sur l'âge pour vérifier que les règles complémentaires de validation fonctionnent aussi.

```python
try:
    bad_age_sample = train.copy()
    bad_age_sample.loc[0, "DAYS_BIRTH"] = -1000
    validate_application_train(bad_age_sample)
except Exception as e:
    print("Erreur métier capturée avec succès :")
    print(str(e)[:1000])
```

## 10. Résumé des règles actuellement contrôlées

Le fichier Pandera contrôle actuellement les éléments suivants :

- unicité de `SK_ID_CURR`
- validité binaire de `TARGET`
- positivité ou nullité de certaines variables financières
- cohérence de `DAYS_BIRTH`
- prise en compte de la valeur spéciale `365243` pour `DAYS_EMPLOYED`
- contrôle des modalités de certaines variables catégorielles
- validation d'un âge dérivé réaliste
- détection d'annuités potentiellement incohérentes par rapport au crédit

Ces règles constituent une première couche de sécurité avant les étapes suivantes du pipeline.
